# Experimento B1: entorno estocástico simple

### Carga de librerías

```python
import random
import numpy as np
import matplotlib.pyplot as plt

import entorno
import agentes
import train
```

### Parámetros de gráficos

```python
plt.rcParams.update({
    'axes.titlesize': 40,     # Title size
    'axes.labelsize': 36,     # X and Y label size
    'xtick.labelsize': 28,    # X tick size
    'ytick.labelsize': 28,    # Y tick size
    'legend.fontsize': 32,    # Legend text size
    'figure.titlesize': 44    # Suptitle size
})
```

### Parámetros del entorno

```python
TAMANOS = {
    'n_estaciones': 4,
    'anclas_por_est': 10,
    'bicis_por_est': 5,    # Estado inicial
    'bicis_por_accion': 2  # Bicis movidas por el agente en c/ accion
}
COSTOS = {
    'accion': 1,
    'accion_invalida': 10,
    'est_vacia': 1,
    'est_llena': 1,
}
TIEMPOS = {
    'viaje': 10,
    'accion': 5,
    'episodio': 8*60,
    'estado': 1,
    'update': 999
}
PROBS = {
    'origen': 4*[0.1],
    'destino': np.array(4*[4*[0.25]])
}
```

### Entrenar agente con $\varepsilon = 0,1$

```python
TRAIN_PARAMS = {
    #'alpha_init': 0.1,
    #'alpha_step': 0.99995,
    #'alpha_end':  0.001,
    'alpha_init': 0.1,
    'alpha_step': 1,
    'alpha_end':  0.1,

    #'eps_init': 0.9,
    #'eps_step': 0.99995,
    #'eps_end':  0.01
    'eps_init': 0.1,
    'eps_step': 1,
    'eps_end':  0.1
}

random.seed(10)

env = entorno.Entorno(TAMANOS, COSTOS, TIEMPOS, PROBS)

print("=== Agente Q-Learning ===")
agente_Q = agentes.QLearningAgent(action_size=env.action_space.n, alpha=TRAIN_PARAMS['alpha_init'], epsilon=TRAIN_PARAMS['eps_init'])
_, metricas_train_test, counters_train_test, best_qtable = train.train_test(env, agente_Q, params=TRAIN_PARAMS, N=10**5, keep_best=True, verbose=False)

# Testear la ultima Q-table
print("\nUltima Q table")
_, metricas_test, counters_test, _ = train.test(env, agente_Q.q_table, N=10**3)

# Testear la mejor Q-table
print("\nMejor Q table")
_, metricas_test, counters_test, _ = train.test(env, best_qtable, N=10**3)
```

```python
train.train_test_plot(metricas_train_test['rewards'])
```
