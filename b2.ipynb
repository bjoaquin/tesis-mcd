### Parámetros del entorno

```python
TAMANOS = {
    'n_estaciones': 4,
    'anclas_por_est': 10,
    'bicis_por_est': 5,    # Estado inicial
    'bicis_por_accion': 2  # Bicis movidas por el agente en c/ accion
}
TIEMPOS = {
    'viaje': 10,
    'accion': 5,
    'episodio': 8*60,
    'estado': 1,
    'update': 999
}
PROBS = {
    'origen': 4*[0.1],
    'destino': np.array(4*[4*[0.25]])
}

TRAIN_PARAMS = {
    'alpha_init': 0.1,
    'alpha_step': 1,
    'alpha_end':  0.1,

    'eps_init': 0.7,
    'eps_step': 1,
    'eps_end':  0.7
}
```

### Entrenar agente con esquema de Costos > Penalizaciones

```python
COSTOS = {
    'accion': 5,
    'accion_invalida': 10,
    'est_vacia': 1,
    'est_llena': 1,
}

random.seed(10)

env = entorno.Entorno(TAMANOS, COSTOS, TIEMPOS, PROBS)

print("=== Agente Q-Learning ===")
agente_Q = agentes.QLearningAgent(action_size=env.action_space.n, alpha=TRAIN_PARAMS['alpha_init'], epsilon=TRAIN_PARAMS['eps_init'])
_, metricas_train_test, counters_train_test, best_qtable = train.train_test(env, agente_Q, params=TRAIN_PARAMS, N=10**5, keep_best=True, verbose=False)

# Testear la ultima Q-table
print("\nUltima Q table")
_, metricas_test, counters_test, _ = train.test(env, agente_Q.q_table, N=10**3)

# Testear la mejor Q-table
print("\nMejor Q table")
_, metricas_test, counters_test, _ = train.test(env, best_qtable, N=10**3)
```

### Intentar explicar la política de Q-Learning

```python
# Entrenar 10 agentes RL y guardar todas sus q-tables para después hacer votacion

q_tables = []

random.seed(505)

env = entorno.Entorno(TAMANOS, COSTOS, TIEMPOS, PROBS)

for i in range(10):
    print(f"Entrenando agente {i+1}/10...")
    agente_Q = agentes.QLearningAgent(action_size=env.action_space.n, alpha=TRAIN_PARAMS['alpha_init'], epsilon=TRAIN_PARAMS['eps_init'])
    _, metricas_train_test, counters_train_test, best_qtable = train.train_test(env, agente_Q, params=TRAIN_PARAMS, N=10**5, keep_best=True, verbose=False)
    q_tables.append(deepcopy(agente_Q.q_table))

# Guardar
q_tables = [dict(q_table) for q_table in q_tables]
with open("q_tables.pkl", "wb") as f:
    pickle.dump(q_tables, f)
# Leer
with open("q_tables.pkl", "rb") as f:
    q_tables = pickle.load(f)
```

```python
# Numero de estados visitados por {algun / todos los} agente(s)

decisiones = [{tuple([int(_) for _ in k[:4]]): np.argmax(v) for k,v in x.items()} for x in q_tables]

estados_visitados = set().union(*(d.keys() for d in decisiones))
print(len(estados_visitados))

estados_visitados_siempre = set(decisiones[0]).intersection(*(d.keys() for d in decisiones[1:]))
print(len(estados_visitados_siempre))
```

```python
# Acuerdo completo (los 10 agentes eligen la misma accion)

decisiones_unanimes = dict()
for estado in estados_visitados_siempre:
    decision = decisiones[0][estado]
    acuerdos = 1
    for q_table in decisiones[1:]:
        if q_table[estado] == decision:
            acuerdos += 1
    if acuerdos == 10:
        decisiones_unanimes[estado] = decision

print(len(decisiones_unanimes))
print(len([v for v in decisiones_unanimes.values() if v == 0]))
```

```python
# Acuerdo mayoritario (al menos 6 agentes eligen la misma accion)

decisiones_casi_unanimes = dict()
for estado in estados_visitados_siempre:
    lista_decisiones = [x[estado] for x in decisiones]
    contador_decisiones = dict(Counter(lista_decisiones))
    max_acuerdos = max(contador_decisiones.values())
    decision_acordada = [k for k,v in contador_decisiones.items() if v == max_acuerdos][0]
    if max_acuerdos >= 6:
        decisiones_casi_unanimes[estado] = decision_acordada

print(len(decisiones_casi_unanimes))
print(Counter(decisiones_casi_unanimes.values()))
```
